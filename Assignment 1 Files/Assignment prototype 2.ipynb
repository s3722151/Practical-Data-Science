```python
#Make the file (before exporting)
#1.We create the mainfile
import pandas as pd
import matplotlib.pyplot as plt
Mainfile = pd.read_csv('Manufacturer.csv', sep=',', decimal='.')
Mainfile

 #This I now can read the file. I can now read the other files.This is from the Week 2 slides - Missing Values
file1 = pd.read_csv('Model.csv',sep=',', decimal='.')
file2 = pd.read_csv('Price.csv',sep=',', decimal='.')
file3 = pd.read_csv('Transmission.csv',sep=',', decimal='.')
file4 = pd.read_csv('Power.csv',sep=',', decimal='.')
file5 = pd.read_csv('Engine CC.csv',sep=',', decimal='.')
file6 = pd.read_csv('Fuel.csv',sep=',', decimal='.')
file7 = pd.read_csv('Female.csv',sep=',', decimal='.')
file8 = pd.read_csv('Male.csv',sep=',', decimal='.')
file9 = pd.read_csv('Unknown.csv',sep=',', decimal='.')
file10 = pd.read_csv('Total.csv',sep=',', decimal='.')

 #Add the files together based on the shared Index.
Mainfile ['Model'] = (file1['Model'])
Mainfile ['Price'] = (file2['Price'])
Mainfile ['Transmission'] = (file3['Transmission'])
Mainfile ['Power'] = (file4['Power'])
Mainfile ['Engine CC'] = (file5['Engine CC'])
Mainfile ['Fuel'] = (file6['Fuel'])
Mainfile ['Female'] = (file7['Female'])
Mainfile ['Male'] = (file8['Male'])
Mainfile ['Unknown'] = (file9['Unknown'])
Mainfile ['Total'] = (file10['Total'])
```

```python
#2.We the clean the file
 #Rename the unamed column then delete it - It is redundant information.
list(Mainfile)
Mainfile.rename(columns = {'Unnamed: 0': 'Index'}, inplace =True)
Mainfile = Mainfile.drop(columns=['Index'])
Mainfile
```

```python
 #Make sure to round all values with decimals to 4 places.
Mainfile['Price'] = Mainfile['Price'].round(2)
Mainfile['Transmission'] = Mainfile['Transmission'].round(2)
Mainfile['Power'] = Mainfile['Power'].round(2)
Mainfile['Engine CC'] = Mainfile['Engine CC'].round(2)
```

```python
 #Getting rid of comma's in our decimal values
Mainfile['Female'] = Mainfile['Female'].str.replace(',', '')
Mainfile['Male'] = Mainfile['Male'].str.replace(',', '')
Mainfile['Unknown'] = Mainfile['Unknown'].str.replace(',', '')
Mainfile['Total'] = Mainfile['Total'].str.replace(',', '')
Mainfile
```

```python
 #We strip white spaces with values that use strings. 
  #This doesn't work on Price, Power and Engine CC. 
  #We notice that we cannot remove reundant spaces of values that have integer values(Price, Transmission, #Power, Engine CC, Female, Male,Unknown, Total)
Mainfile['Manufacturer'] = Mainfile['Manufacturer'].str.strip()
Mainfile['Model'] = Mainfile['Model'].str.strip()
#Mainfile['Price'] = Mainfile['Price'].str.strip()
#Mainfile['Power'] = Mainfile['Power'].str.strip()
#Mainfile['Engine CC'] = Mainfile['Engine CC'].str.strip()
Mainfile['Fuel'] = Mainfile['Fuel'].str.strip()
Mainfile['Male'] = Mainfile['Male'].str.strip()
Mainfile['Female'] = Mainfile['Female'].str.strip()
Mainfile['Unknown'] = Mainfile['Unknown'].str.strip()
Mainfile['Total'] = Mainfile['Total'].str.strip()
Mainfile.shape
```

```python
 #Remove values outside of our range or that are blank.
  #NOTE: I could not execute the conditions for Engine CC
Mainfile.dropna()
Mainfile = Mainfile.drop(Mainfile[(Mainfile.Price <= 0.0)].index)
Mainfile = Mainfile.drop(Mainfile[(Mainfile.Price > 650.0)].index)
# Alternate way: Mainfile = Mainfile.drop(Mainfile[(Mainfile.Price < 0.0) & (Mainfile.Price > 650.0)].index)
Mainfile = Mainfile.drop(Mainfile[(Mainfile.Transmission <= 0.0)].index)
Mainfile = Mainfile.drop(Mainfile[(Mainfile.Transmission > 10.0)].index)
Mainfile = Mainfile.drop(Mainfile[(Mainfile.Power < 0.0)].index)
Mainfile = Mainfile.drop(Mainfile[(Mainfile.Power >500.0)].index)
#Mainfile = Mainfile.drop(Mainfile[(Mainfile.Engine CC< 0.0) & (Mainfile.Engine CC >6500.0)].index)
Mainfile = Mainfile.drop(Mainfile[(Mainfile.Fuel == "")].index)
Mainfile = Mainfile.drop(Mainfile[(Mainfile.Male == "")].index)
Mainfile = Mainfile.drop(Mainfile[(Mainfile.Female == "")].index)
Mainfile = Mainfile.drop(Mainfile[(Mainfile.Unknown == "")].index)
Mainfile = Mainfile.drop(Mainfile[(Mainfile.Total == "")].index)
Mainfile.head(10)
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))
 
# set height of bar
Male = [65505, 79176, 51517, 74157, 47900]
Female = [30221, 13946, 41027, 15889, 39519]
Unknown = [3778, 3968, 4395, 5274, 5986]
 
# Set position of bar on X axis
br1 = np.arange(len(Male))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
 
# Make the plot
plt.bar(br1, Male, color ='r', width = barWidth,
        edgecolor ='grey', label ='Male')
plt.bar(br2, Female, color ='g', width = barWidth,
        edgecolor ='grey', label ='Female')
plt.bar(br3, Unknown, color ='b', width = barWidth,
        edgecolor ='grey', label ='Unknown')
 
# Adding Xticks
plt.xlabel('Car Model', fontweight ='bold', fontsize = 15)
plt.ylabel('No. of people', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(Male))],
        ['Leon', 'XE', 'Fusion', 'Accord', 'MX-5'])

plt.title('Top 10 Car Models ranked 1 to 5')
plt.legend(title = 'Legend')
plt.show()
```
